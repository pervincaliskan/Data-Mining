```python
# =============================================================================
# COVERTYPE VERÄ° SETÄ° ÃœZERÄ°NDE Ã–ZNÄ°TELÄ°K SEÃ‡Ä°MÄ° PROJESÄ°
# FOX OPTIMIZER ALGORÄ°TMASI KULLANARAK - GÃœNCELLENMIÅ VERSÄ°YON
# =============================================================================

"""
Bu proje Covertype veri seti Ã¼zerinde doÄŸadan ilhamlÄ± FOX Optimizer algoritmasÄ±
kullanarak Ã¶znitelik seÃ§imi yapmayÄ± amaÃ§lar.

Proje AdÄ±mlarÄ±:
1. Gerekli kÃ¼tÃ¼phaneleri yÃ¼kle
2. Veri setini indir ve incele
3. FOX Optimizer algoritmasÄ±nÄ± kodla
4. Ã–znitelik seÃ§imi yap
5. FarklÄ± sÄ±nÄ±flandÄ±rÄ±cÄ±larla test et (TÃœM METRÄ°KLER dahil)
6. SonuÃ§larÄ± karÅŸÄ±laÅŸtÄ±r ve raporla

GÃœNCELLEMELER:
- Ã–zgÃ¼llÃ¼k (Specificity) metriÄŸi eklendi
- Performans deÄŸerlendirme fonksiyonu gÃ¼ncellendi
- GÃ¶rselleÅŸtirme kodlarÄ± geliÅŸtirildi
"""

# =============================================================================
# 1. KÃœTÃœPHANALARI YÃœKLEME
# =============================================================================

print("ğŸ“¦ Gerekli kÃ¼tÃ¼phaneler yÃ¼kleniyor...")

# Temel kÃ¼tÃ¼phaneler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import warnings
warnings.filterwarnings('ignore')

# Scikit-learn kÃ¼tÃ¼phaneleri
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                           f1_score, confusion_matrix, classification_report)

# UCI veri seti kÃ¼tÃ¼phanesi
!pip install ucimlrepo

from ucimlrepo import fetch_ucirepo

# Grafiklerin daha gÃ¼zel gÃ¶rÃ¼nmesi iÃ§in
plt.style.use('default')
sns.set_palette("husl")

print("âœ… TÃ¼m kÃ¼tÃ¼phaneler baÅŸarÄ±yla yÃ¼klendi!")

# =============================================================================
# 2. VERÄ° SETÄ°NÄ° Ä°NDÄ°RME VE Ä°NCELEME
# =============================================================================

print("\nğŸŒ² Covertype veri seti indiriliyor...")

# Covertype veri setini UCI Repository'den indir
covertype = fetch_ucirepo(id=31)

# Veri setini Pandas DataFrame'e dÃ¶nÃ¼ÅŸtÃ¼r
X_df = covertype.data.features  # Ã–znitelikler
y_df = covertype.data.targets   # Hedef deÄŸiÅŸken (orman tÃ¼rÃ¼)

print(f"ğŸ“Š Veri seti boyutu: {X_df.shape[0]} satÄ±r x {X_df.shape[1]} Ã¶znitelik")

# Veri seti hakkÄ±nda bilgi
print("\nğŸ“‹ Veri seti meta verileri:")
print(covertype.metadata)

print("\nğŸ“ Ã–znitelik bilgileri:")
print(covertype.variables.head(10))

# Ä°lk 5 satÄ±rÄ± gÃ¶ster
print("\nğŸ‘€ Verinin ilk 5 satÄ±rÄ±:")
print(X_df.head())

print("\nğŸ¯ Hedef deÄŸiÅŸken daÄŸÄ±lÄ±mÄ±:")
print(y_df.value_counts().sort_index())

# =============================================================================
# 3. VERÄ° Ã–N Ä°ÅLEME
# =============================================================================

print("\nğŸ”§ Veri Ã¶n iÅŸleme baÅŸlÄ±yor...")

# Veriyi numpy array'e dÃ¶nÃ¼ÅŸtÃ¼r
X = X_df.values
y = y_df.values.ravel()  # Tek boyutlu hale getir

print(f"Ã–znitelik matrisi boyutu: {X.shape}")
print(f"Hedef deÄŸiÅŸken boyutu: {y.shape}")

# Veri seti Ã§ok bÃ¼yÃ¼k olduÄŸu iÃ§in Ã¶rnekleme yapalÄ±m (iÅŸlemi hÄ±zlandÄ±rmak iÃ§in)
if X.shape[0] > 20000:
    print("âš¡ Veri seti bÃ¼yÃ¼k, Ã¶rnekleme yapÄ±lÄ±yor...")
    np.random.seed(42)
    sample_size = 20000
    indices = np.random.choice(X.shape[0], sample_size, replace=False)
    X = X[indices]
    y = y[indices]
    print(f"Ã–rnekleme sonrasÄ± boyut: {X.shape}")

# SayÄ±sal Ã¶znitelikleri standartlaÅŸtÄ±r (ilk 10 Ã¶znitelik sayÄ±sal)
scaler = StandardScaler()
X_scaled = X.copy()
X_scaled[:, :10] = scaler.fit_transform(X[:, :10])

print("âœ… Veri Ã¶n iÅŸleme tamamlandÄ±!")

# EÄŸitim ve test kÃ¼melerini ayÄ±r
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"EÄŸitim seti boyutu: {X_train.shape}")
print(f"Test seti boyutu: {X_test.shape}")

# =============================================================================
# 4. Ã–ZGÃœLLÃœK METRÄ°ÄÄ° FONKSÄ°YONU
# =============================================================================

def calculate_specificity(y_true, y_pred, average='weighted'):
    """
    Ã–zgÃ¼llÃ¼k (Specificity) hesaplama fonksiyonu

    Specificity = True Negative / (True Negative + False Positive)

    Ã‡ok sÄ±nÄ±flÄ± problemlerde her sÄ±nÄ±f iÃ§in ayrÄ± hesaplanÄ±r ve ortalama alÄ±nÄ±r.
    """
    # Confusion matrix'i al
    cm = confusion_matrix(y_true, y_pred)
    n_classes = len(cm)

    specificities = []
    class_supports = []  # Her sÄ±nÄ±fÄ±n gerÃ§ek Ã¶rnek sayÄ±sÄ±

    # Her sÄ±nÄ±f iÃ§in Ã¶zgÃ¼llÃ¼k hesapla
    for i in range(n_classes):
        # True Negative: i sÄ±nÄ±fÄ± dÄ±ÅŸÄ±ndaki tÃ¼m doÄŸru tahminler
        tn = np.sum(cm) - np.sum(cm[i, :]) - np.sum(cm[:, i]) + cm[i, i]

        # False Positive: i sÄ±nÄ±fÄ± olarak yanlÄ±ÅŸ tahmin edilenler
        fp = np.sum(cm[:, i]) - cm[i, i]

        # Ã–zgÃ¼llÃ¼k hesapla
        if tn + fp == 0:
            specificity = 0.0
        else:
            specificity = tn / (tn + fp)

        specificities.append(specificity)
        class_supports.append(np.sum(cm[i, :]))  # i sÄ±nÄ±fÄ±nÄ±n gerÃ§ek Ã¶rnek sayÄ±sÄ±

    # Ortalama hesaplama tÃ¼rÃ¼
    if average == 'macro':
        return np.mean(specificities)
    elif average == 'weighted':
        # SÄ±nÄ±f bÃ¼yÃ¼klÃ¼klerine gÃ¶re aÄŸÄ±rlÄ±klÄ± ortalama
        total_support = sum(class_supports)
        weights = [support / total_support for support in class_supports]
        return np.average(specificities, weights=weights)
    elif average == 'micro':
        # Micro average (genel TN ve FP hesabÄ±)
        total_tn = sum([np.sum(cm) - np.sum(cm[i, :]) - np.sum(cm[:, i]) + cm[i, i]
                       for i in range(n_classes)]) / n_classes
        total_fp = sum([np.sum(cm[:, i]) - cm[i, i] for i in range(n_classes)]) / n_classes
        return total_tn / (total_tn + total_fp) if (total_tn + total_fp) > 0 else 0.0
    else:
        # SÄ±nÄ±f bazÄ±nda deÄŸerler dÃ¶ndÃ¼r
        return specificities

# =============================================================================
# 5. FOX OPTIMIZER ALGORÄ°TMASINI KODLAMA
# =============================================================================

print("\nğŸ¦Š FOX Optimizer algoritmasÄ± kodlanÄ±yor...")

class FOXOptimizer:
    """
    FOX Optimizer AlgoritmasÄ±

    Bu algoritma tilkilerin avlanma davranÄ±ÅŸlarÄ±ndan esinlenir:
    - Tilkiler avlarÄ±nÄ± takip eder (exploration - keÅŸif)
    - AvlarÄ±nÄ± kuÅŸatÄ±r (exploitation - sÃ¶mÃ¼rÃ¼)
    - En iyi avÄ± bulana kadar bu stratejileri tekrarlar

    Ã–znitelik seÃ§imi iÃ§in:
    - Her tilki bir Ã¶znitelik alt kÃ¼mesini temsil eder
    - En iyi sÄ±nÄ±flandÄ±rma performansÄ±nÄ± veren alt kÃ¼meyi arar
    """

    def __init__(self, n_foxes=15, max_iter=15, p_pursuit=0.3, p_encircle=0.4):
        """
        Parametreler:
        - n_foxes: Tilki sayÄ±sÄ± (popÃ¼lasyon bÃ¼yÃ¼klÃ¼ÄŸÃ¼)
        - max_iter: Maksimum iterasyon sayÄ±sÄ±
        - p_pursuit: Takip etme olasÄ±lÄ±ÄŸÄ±
        - p_encircle: KuÅŸatma olasÄ±lÄ±ÄŸÄ±
        """
        self.n_foxes = n_foxes
        self.max_iter = max_iter
        self.p_pursuit = p_pursuit
        self.p_encircle = p_encircle
        self.best_solution = None
        self.best_fitness = 0.0
        self.best_subset = None
        self.fitness_history = []

    def initialize_population(self, n_features):
        """
        Tilki popÃ¼lasyonunu baÅŸlat
        Her tilki binary bir vektÃ¶r (0: Ã¶znitelik seÃ§ilmedi, 1: seÃ§ildi)
        """
        population = np.random.randint(0, 2, size=(self.n_foxes, n_features))

        # Her tilkinin en az bir Ã¶znitelik seÃ§tiÄŸinden emin ol
        for i in range(self.n_foxes):
            if np.sum(population[i]) == 0:
                population[i, np.random.randint(0, n_features)] = 1

        return population

    def calculate_fitness(self, solution, X, y):
        """
        Bir Ã§Ã¶zÃ¼mÃ¼n fitness deÄŸerini hesapla

        Fitness = SÄ±nÄ±flandÄ±rma DoÄŸruluÄŸu - Ã–znitelik CezasÄ±

        Bu ÅŸekilde hem yÃ¼ksek doÄŸruluk hem de az Ã¶znitelik kullanmayÄ± hedefleriz
        """
        # HiÃ§ Ã¶znitelik seÃ§ilmediyse 0 dÃ¶ndÃ¼r
        if np.sum(solution) == 0:
            return 0.0

        # SeÃ§ilen Ã¶znitelikleri bul
        selected_features = np.where(solution == 1)[0]

        # SeÃ§ilen Ã¶zniteliklerle veri alt kÃ¼mesi oluÅŸtur
        X_selected = X[:, selected_features]

        # K-NN sÄ±nÄ±flandÄ±rÄ±cÄ± ile Ã§apraz doÄŸrulama yap
        knn = KNeighborsClassifier(n_neighbors=5)
        # Using a smaller cv value to reduce execution time
        scores = cross_val_score(knn, X_selected, y, cv=2, scoring='accuracy')

        # Ortalama doÄŸruluk
        accuracy = np.mean(scores)

        # Ã–znitelik sayÄ±sÄ± cezasÄ± (Ã§ok fazla Ã¶znitelik seÃ§meyi Ã¶nlemek iÃ§in)
        feature_ratio = np.sum(solution) / len(solution)
        penalty = 0.01 * feature_ratio

        fitness = accuracy - penalty
        return max(0, fitness)  # Negatif deÄŸer olmasÄ±nÄ± Ã¶nle

    def pursue_prey(self, fox, best_fox):
        """
        Tilkinin en iyi tilkiyi takip etmesi (keÅŸif fazÄ±)
        """
        new_fox = fox.copy()
        for i in range(len(fox)):
            if np.random.random() < self.p_pursuit:
                new_fox[i] = best_fox[i]
        return new_fox

    def encircle_prey(self, fox):
        """
        Tilkinin mevcut pozisyonunu deÄŸiÅŸtirmesi (sÃ¶mÃ¼rÃ¼ fazÄ±)
        """
        new_fox = fox.copy()
        for i in range(len(fox)):
            if np.random.random() < self.p_encircle:
                new_fox[i] = 1 - new_fox[i]  # Bit Ã§evirme

        # En az bir Ã¶znitelik seÃ§ildiÄŸinden emin ol
        if np.sum(new_fox) == 0:
            new_fox[np.random.randint(0, len(new_fox))] = 1

        return new_fox

    def fit(self, X, y):
        """
        FOX Optimizer algoritmasÄ±nÄ± Ã§alÄ±ÅŸtÄ±r
        """
        print(f"ğŸ¦Š FOX Optimizer baÅŸlÄ±yor: {self.n_foxes} tilki, {self.max_iter} iterasyon")

        n_features = X.shape[1]

        # 1. PopÃ¼lasyonu baÅŸlat
        population = self.initialize_population(n_features)
        fitness_values = np.zeros(self.n_foxes)

        # 2. Ä°lk fitness deÄŸerlerini hesapla
        print("ğŸ§® Ä°lk fitness deÄŸerleri hesaplanÄ±yor...")
        for i in range(self.n_foxes):
            fitness_values[i] = self.calculate_fitness(population[i], X, y)

        # 3. En iyi tilkiyi bul
        best_idx = np.argmax(fitness_values)
        self.best_solution = population[best_idx].copy()
        self.best_fitness = fitness_values[best_idx]

        print(f"Ä°lk en iyi fitness: {self.best_fitness:.4f}")

        # 4. Ana algoritma dÃ¶ngÃ¼sÃ¼
        for iteration in range(self.max_iter):
            print(f"\nğŸ”„ Ä°terasyon {iteration + 1}/{self.max_iter}")

            # Her tilki iÃ§in avlanma stratejisi uygula
            for i in range(self.n_foxes):
                current_fox = population[i].copy()
                current_fitness = fitness_values[i]

                # Takip etme stratejisi
                fox_pursuit = self.pursue_prey(current_fox, self.best_solution)
                fitness_pursuit = self.calculate_fitness(fox_pursuit, X, y)

                # KuÅŸatma stratejisi
                fox_encircle = self.encircle_prey(current_fox)
                fitness_encircle = self.calculate_fitness(fox_encircle, X, y)

                # En iyi stratejiyi seÃ§
                if fitness_pursuit > max(fitness_encircle, current_fitness):
                    population[i] = fox_pursuit
                    fitness_values[i] = fitness_pursuit
                elif fitness_encircle > current_fitness:
                    population[i] = fox_encircle
                    fitness_values[i] = fitness_encircle

                # Global en iyi Ã§Ã¶zÃ¼mÃ¼ gÃ¼ncelle
                if fitness_values[i] > self.best_fitness:
                    self.best_solution = population[i].copy()
                    self.best_fitness = fitness_values[i]

            # Ä°lerleme bilgisi
            selected_count = np.sum(self.best_solution)
            avg_fitness = np.mean(fitness_values)

            print(f"En iyi fitness: {self.best_fitness:.4f}")
            print(f"Ortalama fitness: {avg_fitness:.4f}")
            print(f"SeÃ§ilen Ã¶znitelik sayÄ±sÄ±: {selected_count}/{n_features}")

            # Fitness geÃ§miÅŸini kaydet
            self.fitness_history.append(self.best_fitness)

        # 5. Son sonuÃ§larÄ± kaydet
        self.best_subset = np.where(self.best_solution == 1)[0]

        print(f"\nâœ… FOX Optimizer tamamlandÄ±!")
        print(f"Final en iyi fitness: {self.best_fitness:.4f}")
        print(f"SeÃ§ilen Ã¶znitelik sayÄ±sÄ±: {len(self.best_subset)}")

        return self

    def transform(self, X):
        """
        Veriyi seÃ§ilen Ã¶zniteliklerle dÃ¶nÃ¼ÅŸtÃ¼r
        """
        if self.best_subset is None:
            raise ValueError("Ã–nce fit() metodunu Ã§aÄŸÄ±rÄ±n!")
        return X[:, self.best_subset]

    def plot_convergence(self):
        """
        AlgoritmanÄ±n yakÄ±nsama grafiÄŸini Ã§iz
        """
        plt.figure(figsize=(10, 6))
        plt.plot(range(1, len(self.fitness_history) + 1), self.fitness_history,
                'b-o', linewidth=2, markersize=6)
        plt.title('FOX Optimizer YakÄ±nsama GrafiÄŸi', fontsize=14, fontweight='bold')
        plt.xlabel('Ä°terasyon', fontsize=12)
        plt.ylabel('En Ä°yi Fitness DeÄŸeri', fontsize=12)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

print("âœ… FOX Optimizer sÄ±nÄ±fÄ± baÅŸarÄ±yla tanÄ±mlandÄ±!")

# =============================================================================
# 6. FOX OPTIMIZER Ä°LE Ã–ZNÄ°TELÄ°K SEÃ‡Ä°MÄ°
# =============================================================================

print("\nğŸ¯ FOX Optimizer ile Ã¶znitelik seÃ§imi baÅŸlÄ±yor...")

# FOX Optimizer nesnesini oluÅŸtur
fox_optimizer = FOXOptimizer(
    n_foxes=15,        # 15 tilki kullan
    max_iter=15,       # 15 iterasyon
    p_pursuit=0.3,     # %30 takip etme olasÄ±lÄ±ÄŸÄ±
    p_encircle=0.4     # %40 kuÅŸatma olasÄ±lÄ±ÄŸÄ±
)

# AlgoritmanÄ±n Ã§alÄ±ÅŸma zamanÄ±nÄ± Ã¶lÃ§
start_time = time.time()

# Ã–znitelik seÃ§imini yap
fox_optimizer.fit(X_train, y_train)

end_time = time.time()
print(f"\nâ±ï¸ Toplam Ã§alÄ±ÅŸma sÃ¼resi: {end_time - start_time:.2f} saniye")

# SeÃ§ilen Ã¶znitelikleri gÃ¶ster
selected_features = fox_optimizer.best_subset
print(f"\nğŸ¯ SeÃ§ilen Ã¶znitelikler ({len(selected_features)} adet):")
print(f"Ä°ndeksler: {selected_features}")

# Ã–znitelik isimlerini gÃ¶ster
feature_names = X_df.columns.tolist()
selected_feature_names = [feature_names[i] for i in selected_features]
print(f"\nÃ–znitelik isimleri:")
for i, name in enumerate(selected_feature_names):
    print(f"{i+1:2d}. {name}")

# YakÄ±nsama grafiÄŸini Ã§iz
fox_optimizer.plot_convergence()

# =============================================================================
# 7. SEÃ‡Ä°LEN Ã–ZNÄ°TELÄ°KLERÄ° GÃ–RSELLEÅTÄ°RME
# =============================================================================

print("\nğŸ“Š SeÃ§ilen Ã¶znitelikler gÃ¶rselleÅŸtiriliyor...")

# Ã–znitelik seÃ§imi sonuÃ§larÄ±nÄ± gÃ¶rselleÅŸtir
plt.figure(figsize=(15, 8))

# TÃ¼m Ã¶znitelikler iÃ§in binary gÃ¶sterim
feature_selection = np.zeros(len(feature_names))
feature_selection[selected_features] = 1

# Bar grafiÄŸi
bars = plt.bar(range(len(feature_names)), feature_selection,
               color=['green' if x == 1 else 'lightgray' for x in feature_selection])

plt.title('FOX Optimizer ile SeÃ§ilen Ã–znitelikler', fontsize=16, fontweight='bold')
plt.xlabel('Ã–znitelik Ä°ndeksi', fontsize=12)
plt.ylabel('SeÃ§ildi (1) / SeÃ§ilmedi (0)', fontsize=12)
plt.ylim(-0.1, 1.1)

# SeÃ§ilen Ã¶znitelikleri vurgula
for i, bar in enumerate(bars):
    if feature_selection[i] == 1:
        bar.set_color('darkgreen')
        bar.set_alpha(0.8)

        # Ã–znitelik ismini yazdÄ±r
        plt.text(i, 1.05, f'{i}', ha='center', va='bottom',
                fontsize=8, rotation=45, color='darkgreen', fontweight='bold')

plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

# SeÃ§ilen Ã¶znitelik tÃ¼rlerinin daÄŸÄ±lÄ±mÄ±
print("\nğŸ“ˆ SeÃ§ilen Ã¶znitelik tÃ¼rleri:")

# Ã–znitelik tÃ¼rlerini kategorize et
continuous_features = []  # SÃ¼rekli Ã¶znitelikler (ilk 10)
wilderness_features = []  # Wilderness Area (50-53)
soil_features = []        # Soil Type (10-49)

for idx in selected_features:
    if idx < 10:
        continuous_features.append(feature_names[idx])
    elif 10 <= idx < 50:
        soil_features.append(feature_names[idx])
    else:
        wilderness_features.append(feature_names[idx])

print(f"SÃ¼rekli Ã¶znitelikler: {len(continuous_features)} adet")
print(f"Toprak tipi Ã¶znitelikleri: {len(soil_features)} adet")
print(f"Wilderness area Ã¶znitelikleri: {len(wilderness_features)} adet")

# Pasta grafiÄŸi
categories = ['SÃ¼rekli', 'Toprak Tipi', 'Wilderness Area']
sizes = [len(continuous_features), len(soil_features), len(wilderness_features)]
colors = ['lightblue', 'lightcoral', 'lightgreen']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=categories, colors=colors, autopct='%1.1f%%',
        startangle=90, textprops={'fontsize': 12})
plt.title('SeÃ§ilen Ã–zniteliklerin TÃ¼r DaÄŸÄ±lÄ±mÄ±', fontsize=14, fontweight='bold')
plt.axis('equal')
plt.show()

# =============================================================================
# 8. GÃœNCELLENMIÅ PERFORMANS DEÄERLENDÄ°RME FONKSÄ°YONU (TÃœM METRÄ°KLER)
# =============================================================================

def evaluate_classifier_complete(clf, X_train, X_test, y_train, y_test, name, feature_type):
    """
    SÄ±nÄ±flandÄ±rÄ±cÄ±nÄ±n performansÄ±nÄ± deÄŸerlendir (TÃœM GEREKLÄ° METRÄ°KLER DAHÄ°L)
    """
    print(f"\n--- {name} ({feature_type}) ---")

    # EÄŸitim sÃ¼resi Ã¶lÃ§Ã¼mÃ¼
    start_time = time.time()
    clf.fit(X_train, y_train)
    training_time = time.time() - start_time

    # Tahmin
    y_pred = clf.predict(X_test)

    # PERFORMANS METRÄ°KLERÄ° (TÃœM GEREKLÄ° METRÄ°KLER)
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)
    specificity = calculate_specificity(y_test, y_pred, average='weighted')

    print(f"â±ï¸  EÄŸitim sÃ¼resi: {training_time:.3f} saniye")
    print(f"ğŸ¯ DoÄŸruluk (Accuracy): {accuracy:.4f}")
    print(f"ğŸ” Hassasiyet (Precision): {precision:.4f}")
    print(f"ğŸ“Š DuyarlÄ±lÄ±k (Recall/Sensitivity): {recall:.4f}")
    print(f"âš–ï¸  F1-Score: {f1:.4f}")
    print(f"ğŸ›¡ï¸  Ã–zgÃ¼llÃ¼k (Specificity): {specificity:.4f}")

    return {
        'classifier': name,
        'feature_type': feature_type,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'specificity': specificity,
        'training_time': training_time,
        'n_features': X_train.shape[1]
    }

# =============================================================================
# 9. SINIFLANDIRICILAR Ä°LE PERFORMANS DEÄERLENDÄ°RMESÄ° (TÃœM METRÄ°KLER)
# =============================================================================

print("\nğŸ¤– FarklÄ± sÄ±nÄ±flandÄ±rÄ±cÄ±larla performans deÄŸerlendirmesi (TÃœM METRÄ°KLER)...")

# SeÃ§ilen Ã¶zniteliklerle eÄŸitim ve test verilerini dÃ¶nÃ¼ÅŸtÃ¼r
X_train_selected = fox_optimizer.transform(X_train)
X_test_selected = fox_optimizer.transform(X_test)

print(f"Orijinal Ã¶znitelik sayÄ±sÄ±: {X_train.shape[1]}")
print(f"SeÃ§ilen Ã¶znitelik sayÄ±sÄ±: {X_train_selected.shape[1]}")

# SÄ±nÄ±flandÄ±rÄ±cÄ±larÄ± tanÄ±mla
classifiers = {
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5),
    'Decision Tree': DecisionTreeClassifier(random_state=42, max_depth=10),
    'Naive Bayes': GaussianNB(),
    'Random Forest': RandomForestClassifier(n_estimators=50, random_state=42, max_depth=10),
    'Support Vector Machine': SVC(kernel='rbf', random_state=42, C=1.0)
}

# SonuÃ§larÄ± saklamak iÃ§in liste
results = []

print("\n" + "="*60)
print("PERFORMANS DEÄERLENDÄ°RMESÄ° (TÃœM METRÄ°KLER)")
print("="*60)

# Her sÄ±nÄ±flandÄ±rÄ±cÄ± iÃ§in hem tÃ¼m Ã¶znitelikler hem de seÃ§ilen Ã¶znitelikler ile test et
for name, clf in classifiers.items():

    # TÃ¼m Ã¶zniteliklerle test et
    result_all = evaluate_classifier_complete(
        clf, X_train, X_test, y_train, y_test, name, "TÃ¼m Ã–znitelikler"
    )
    results.append(result_all)

    # SeÃ§ilen Ã¶zniteliklerle test et
    result_selected = evaluate_classifier_complete(
        clf, X_train_selected, X_test_selected, y_train, y_test, name, "SeÃ§ilen Ã–znitelikler"
    )
    results.append(result_selected)

# =============================================================================
# 10. SONUÃ‡LARI KARÅILAÅTIRMA VE GÃ–RSELLEÅTÄ°RME (GÃœNCELLENMIÅ)
# =============================================================================

print("\nğŸ“Š SonuÃ§lar karÅŸÄ±laÅŸtÄ±rÄ±lÄ±yor ve gÃ¶rselleÅŸtiriliyor...")

# SonuÃ§larÄ± DataFrame'e dÃ¶nÃ¼ÅŸtÃ¼r
results_df = pd.DataFrame(results)

# SonuÃ§larÄ± pivot table olarak dÃ¼zenle
metrics = ['accuracy', 'precision', 'recall', 'f1_score', 'specificity', 'training_time']

# Her metrik iÃ§in karÅŸÄ±laÅŸtÄ±rma grafiÄŸi
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.ravel()

classifier_names = list(classifiers.keys())
n_classifiers = len(classifier_names)

for i, metric in enumerate(metrics):
    ax = axes[i]

    # TÃ¼m Ã¶znitelikler iÃ§in deÄŸerler
    all_features_values = []
    selected_features_values = []

    for clf_name in classifier_names:
        all_val = results_df[(results_df['classifier'] == clf_name) &
                           (results_df['feature_type'] == 'TÃ¼m Ã–znitelikler')][metric].values[0]
        selected_val = results_df[(results_df['classifier'] == clf_name) &
                                (results_df['feature_type'] == 'SeÃ§ilen Ã–znitelikler')][metric].values[0]

        all_features_values.append(all_val)
        selected_features_values.append(selected_val)

    # Bar grafiÄŸi
    x = np.arange(n_classifiers)
    width = 0.35

    bars1 = ax.bar(x - width/2, all_features_values, width,
                   label='TÃ¼m Ã–znitelikler', alpha=0.8, color='lightblue')
    bars2 = ax.bar(x + width/2, selected_features_values, width,
                   label='SeÃ§ilen Ã–znitelikler', alpha=0.8, color='lightcoral')

    # Grafik dÃ¼zenlemeleri
    ax.set_xlabel('SÄ±nÄ±flandÄ±rÄ±cÄ±')
    ax.set_ylabel(metric.replace('_', ' ').title())
    ax.set_title(f'{metric.replace("_", " ").title()} KarÅŸÄ±laÅŸtÄ±rmasÄ±')
    ax.set_xticks(x)
    ax.set_xticklabels([name.split()[0] for name in classifier_names], rotation=45)
    ax.legend()
    ax.grid(True, alpha=0.3)

    # DeÄŸerleri bar Ã¼zerinde gÃ¶ster
    for bar in bars1:
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.3f}', ha='center', va='bottom', fontsize=8)

    for bar in bars2:
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.3f}', ha='center', va='bottom', fontsize=8)

plt.tight_layout()
plt.show()

# =============================================================================
# 11. Ã–ZET RAPOR
# =============================================================================

print("\n" + "="*60)
print("PROJE Ã–ZET RAPORU")
print("="*60)

print(f"\nğŸ“Š VERÄ° SETÄ° BÄ°LGÄ°LERÄ°:")
print(f"   â€¢ Toplam Ã¶rnek sayÄ±sÄ±: {len(y)}")
print(f"   â€¢ Toplam Ã¶znitelik sayÄ±sÄ±: {X.shape[1]}")
print(f"   â€¢ SÄ±nÄ±f sayÄ±sÄ±: {len(np.unique(y))}")

print(f"\nğŸ¦Š FOX OPTIMIZER SONUÃ‡LARI:")
print(f"   â€¢ SeÃ§ilen Ã¶znitelik sayÄ±sÄ±: {len(selected_features)}")
print(f"   â€¢ Ã–znitelik azaltma oranÄ±: %{(1 - len(selected_features)/X.shape[1])*100:.1f}")
print(f"   â€¢ En iyi fitness deÄŸeri: {fox_optimizer.best_fitness:.4f}")

print(f"\nğŸ† EN Ä°YÄ° PERFORMANS GÃ–STERENLERÄ°:")

# En iyi doÄŸruluk
best_accuracy = results_df.loc[results_df['accuracy'].idxmax()]
print(f"\n   ğŸ“ En YÃ¼ksek DoÄŸruluk:")
print(f"      {best_accuracy['classifier']} ({best_accuracy['feature_type']})")
print(f"      DoÄŸruluk: {best_accuracy['accuracy']:.4f}")

# En iyi F1-score
best_f1 = results_df.loc[results_df['f1_score'].idxmax()]
print(f"\n   ğŸ“ En YÃ¼ksek F1-Score:")
print(f"      {best_f1['classifier']} ({best_f1['feature_type']})")
print(f"      F1-Score: {best_f1['f1_score']:.4f}")

# En iyi Ã–zgÃ¼llÃ¼k
best_specificity = results_df.loc[results_df['specificity'].idxmax()]
print(f"\n   ğŸ“ En YÃ¼ksek Ã–zgÃ¼llÃ¼k:")
print(f"      {best_specificity['classifier']} ({best_specificity['feature_type']})")
print(f"      Ã–zgÃ¼llÃ¼k: {best_specificity['specificity']:.4f}")

# En hÄ±zlÄ± eÄŸitim
fastest_training = results_df.loc[results_df['training_time'].idxmin()]
print(f"\n   ğŸ“ En HÄ±zlÄ± EÄŸitim:")
print(f"      {fastest_training['classifier']} ({fastest_training['feature_type']})")
print(f"      SÃ¼re: {fastest_training['training_time']:.3f} saniye")

print(f"\nğŸ’¡ Ã–NEMLÄ° BULGULAR:")

# SeÃ§ilen Ã¶zniteliklerle performans karÅŸÄ±laÅŸtÄ±rmasÄ±
all_features_avg = results_df[results_df['feature_type'] == 'TÃ¼m Ã–znitelikler']['accuracy'].mean()
selected_features_avg = results_df[results_df['feature_type'] == 'SeÃ§ilen Ã–znitelikler']['accuracy'].mean()

print(f"   â€¢ TÃ¼m Ã¶zniteliklerle ortalama doÄŸruluk: {all_features_avg:.4f}")
print(f"   â€¢ SeÃ§ilen Ã¶zniteliklerle ortalama doÄŸruluk: {selected_features_avg:.4f}")

if selected_features_avg > all_features_avg:
    print(f"   âœ… FOX Optimizer ile seÃ§ilen Ã¶znitelikler daha iyi performans gÃ¶sterdi!")
    print(f"      Ä°yileÅŸtirme: %{((selected_features_avg - all_features_avg) / all_features_avg) * 100:.2f}")
else:
    print(f"   âš ï¸  Ã–znitelik seÃ§imi performansÄ± biraz dÃ¼ÅŸÃ¼rdÃ¼, ancak hesaplama maliyetini azalttÄ±.")

# EÄŸitim sÃ¼resi karÅŸÄ±laÅŸtÄ±rmasÄ±
all_features_time = results_df[results_df['feature_type'] == 'TÃ¼m Ã–znitelikler']['training_time'].mean()
selected_features_time = results_df[results_df['feature_type'] == 'SeÃ§ilen Ã–znitelikler']['training_time'].mean()

print(f"\n   â±ï¸  Ortalama eÄŸitim sÃ¼releri:")
print(f"      TÃ¼m Ã¶znitelikler: {all_features_time:.3f} saniye")
print(f"      SeÃ§ilen Ã¶znitelikler: {selected_features_time:.3f} saniye")
print(f"      HÄ±zlanma: %{((all_features_time - selected_features_time) / all_features_time) * 100:.1f}")

# Ã–zgÃ¼llÃ¼k performansÄ±
all_features_spec = results_df[results_df['feature_type'] == 'TÃ¼m Ã–znitelikler']['specificity'].mean()
selected_features_spec = results_df[results_df['feature_type'] == 'SeÃ§ilen Ã–znitelikler']['specificity'].mean()

print(f"\n   ğŸ›¡ï¸  Ortalama Ã¶zgÃ¼llÃ¼k deÄŸerleri:")
print(f"      TÃ¼m Ã¶znitelikler: {all_features_spec:.4f}")
print(f"      SeÃ§ilen Ã¶znitelikler: {selected_features_spec:.4f}")

# =============================================================================
# 12. DETAYLI SONUÃ‡ TABLOSU
# =============================================================================

print(f"\nğŸ“‹ DETAYLI SONUÃ‡ TABLOSU:")
print("="*60)

# SonuÃ§larÄ± dÃ¼zenli tablo halinde gÃ¶ster
display_df = results_df.round(4)
display_df = display_df.sort_values(['classifier', 'feature_type'])

# SÃ¼tun isimlerini TÃ¼rkÃ§eleÅŸtir
display_df_turkish = display_df.copy()
display_df_turkish.columns = ['SÄ±nÄ±flandÄ±rÄ±cÄ±', 'Ã–znitelik TÃ¼rÃ¼', 'DoÄŸruluk',
                             'Hassasiyet', 'DuyarlÄ±lÄ±k', 'F1-PuanÄ±', 'Ã–zgÃ¼llÃ¼k',
                             'EÄŸitim SÃ¼resi', 'Ã–znitelik SayÄ±sÄ±']

print(display_df_turkish.to_string(index=False))

# =============================================================================
# 13. SEÃ‡Ä°LEN Ã–ZNÄ°TELÄ°KLERÄ°N DETAYLI ANALÄ°ZÄ°
# =============================================================================

print(f"\nğŸ” SEÃ‡Ä°LEN Ã–ZNÄ°TELÄ°KLERÄ°N DETAYLI ANALÄ°ZÄ°:")
print("="*60)

# SeÃ§ilen Ã¶zniteliklerin istatistikleri
print(f"\nğŸ“Š SeÃ§ilen Ã–znitelik Ä°statistikleri:")
selected_data = X_train_selected
feature_stats = []

for i, feature_idx in enumerate(selected_features):
    feature_name = feature_names[feature_idx]
    feature_data = X_train[:, feature_idx]

    stats = {
        'Ä°ndeks': feature_idx,
        'Ä°sim': feature_name,
        'Ortalama': np.mean(feature_data),
        'Std Sapma': np.std(feature_data),
        'Min': np.min(feature_data),
        'Max': np.max(feature_data)
    }
    feature_stats.append(stats)

feature_stats_df = pd.DataFrame(feature_stats)
print(feature_stats_df.round(3).to_string(index=False))

# =============================================================================
# 14. GÃ–RSELLEÅTÄ°RME: PERFORMANS RADAR CHART
# =============================================================================

print(f"\nğŸ“Š Radar chart ile performans gÃ¶rselleÅŸtirmesi...")

# En iyi performans gÃ¶steren sÄ±nÄ±flandÄ±rÄ±cÄ±larÄ± seÃ§
top_classifiers = ['Random Forest', 'K-Nearest Neighbors', 'Decision Tree']

# Radar chart iÃ§in veri hazÄ±rla
metrics_for_radar = ['accuracy', 'precision', 'recall', 'f1_score', 'specificity']
angles = np.linspace(0, 2 * np.pi, len(metrics_for_radar), endpoint=False).tolist()
angles += angles[:1]  # DÃ¶ngÃ¼yÃ¼ tamamla

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7), subplot_kw=dict(projection='polar'))

colors = ['blue', 'red', 'green']

# TÃ¼m Ã¶znitelikler iÃ§in radar chart
ax1.set_title('TÃ¼m Ã–znitelikler ile Performans', y=1.08, fontsize=14, fontweight='bold')
for i, clf_name in enumerate(top_classifiers):
    values = []
    for metric in metrics_for_radar:
        value = results_df[(results_df['classifier'] == clf_name) &
                          (results_df['feature_type'] == 'TÃ¼m Ã–znitelikler')][metric].values[0]
        values.append(value)
    values += values[:1]  # DÃ¶ngÃ¼yÃ¼ tamamla

    ax1.plot(angles, values, 'o-', linewidth=2, label=clf_name, color=colors[i])
    ax1.fill(angles, values, alpha=0.25, color=colors[i])

ax1.set_xticks(angles[:-1])
ax1.set_xticklabels([m.replace('_', '\n').title() for m in metrics_for_radar])
ax1.set_ylim(0, 1)
ax1.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
ax1.grid(True)

# SeÃ§ilen Ã¶znitelikler iÃ§in radar chart
ax2.set_title('SeÃ§ilen Ã–znitelikler ile Performans', y=1.08, fontsize=14, fontweight='bold')
for i, clf_name in enumerate(top_classifiers):
    values = []
    for metric in metrics_for_radar:
        value = results_df[(results_df['classifier'] == clf_name) &
                          (results_df['feature_type'] == 'SeÃ§ilen Ã–znitelikler')][metric].values[0]
        values.append(value)
    values += values[:1]  # DÃ¶ngÃ¼yÃ¼ tamamla

    ax2.plot(angles, values, 'o-', linewidth=2, label=clf_name, color=colors[i])
    ax2.fill(angles, values, alpha=0.25, color=colors[i])

ax2.set_xticks(angles[:-1])
ax2.set_xticklabels([m.replace('_', '\n').title() for m in metrics_for_radar])
ax2.set_ylim(0, 1)
ax2.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
ax2.grid(True)

plt.tight_layout()
plt.show()

# =============================================================================
# 15. CONFUSION MATRIX GÃ–RSELLEÅTÄ°RMESÄ°
# =============================================================================

print(f"\nğŸ¯ En iyi performans gÃ¶steren modelin confusion matrix'i...")

# En iyi F1-score'a sahip modeli bul
best_model_info = results_df.loc[results_df['f1_score'].idxmax()]
best_classifier_name = best_model_info['classifier']
best_feature_type = best_model_info['feature_type']

print(f"En iyi model: {best_classifier_name} ({best_feature_type})")

# En iyi modeli yeniden eÄŸit ve confusion matrix oluÅŸtur
if best_feature_type == 'SeÃ§ilen Ã–znitelikler':
    X_train_best = X_train_selected
    X_test_best = X_test_selected
else:
    X_train_best = X_train
    X_test_best = X_test

# Modeli seÃ§ ve eÄŸit
if 'Random Forest' in best_classifier_name:
    best_clf = RandomForestClassifier(n_estimators=50, random_state=42, max_depth=10)
elif 'K-Nearest' in best_classifier_name:
    best_clf = KNeighborsClassifier(n_neighbors=5)
elif 'Decision Tree' in best_classifier_name:
    best_clf = DecisionTreeClassifier(random_state=42, max_depth=10)
elif 'Naive Bayes' in best_classifier_name:
    best_clf = GaussianNB()
else:  # SVM
    best_clf = SVC(kernel='rbf', random_state=42, C=1.0)

best_clf.fit(X_train_best, y_train)
y_pred_best = best_clf.predict(X_test_best)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred_best)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=[f'SÄ±nÄ±f {i}' for i in range(1, 8)],
            yticklabels=[f'SÄ±nÄ±f {i}' for i in range(1, 8)])
plt.title(f'Confusion Matrix - {best_classifier_name}\n({best_feature_type})',
          fontsize=14, fontweight='bold')
plt.xlabel('Tahmin Edilen SÄ±nÄ±f', fontsize=12)
plt.ylabel('GerÃ§ek SÄ±nÄ±f', fontsize=12)
plt.tight_layout()
plt.show()

# SÄ±nÄ±f bazÄ±nda performans raporu
print(f"\nğŸ“Š SÄ±nÄ±f bazÄ±nda detaylÄ± performans raporu:")
class_report = classification_report(y_test, y_pred_best,
                                   target_names=[f'SÄ±nÄ±f {i}' for i in range(1, 8)])
print(class_report)

# =============================================================================
# 16. METRÄ°K BAZINDA DETAYLI KARÅILAÅTIRMA
# =============================================================================

print(f"\nğŸ“ˆ Metrik bazÄ±nda detaylÄ± karÅŸÄ±laÅŸtÄ±rma:")
print("="*60)

# Her metrik iÃ§in en iyi performans gÃ¶steren model
metrics_analysis = ['accuracy', 'precision', 'recall', 'f1_score', 'specificity']

for metric in metrics_analysis:
    best_metric = results_df.loc[results_df[metric].idxmax()]
    print(f"\nğŸ† En Ä°yi {metric.replace('_', ' ').title()}:")
    print(f"   Model: {best_metric['classifier']}")
    print(f"   Ã–znitelik: {best_metric['feature_type']}")
    print(f"   DeÄŸer: {best_metric[metric]:.4f}")

    # Bu metrik iÃ§in tÃ¼m vs seÃ§ilen Ã¶znitelik karÅŸÄ±laÅŸtÄ±rmasÄ±
    all_avg = results_df[results_df['feature_type'] == 'TÃ¼m Ã–znitelikler'][metric].mean()
    selected_avg = results_df[results_df['feature_type'] == 'SeÃ§ilen Ã–znitelikler'][metric].mean()

    print(f"   TÃ¼m Ã¶znitelik ortalamasÄ±: {all_avg:.4f}")
    print(f"   SeÃ§ilen Ã¶znitelik ortalamasÄ±: {selected_avg:.4f}")

    if selected_avg > all_avg:
        improvement = ((selected_avg - all_avg) / all_avg) * 100
        print(f"   âœ… Ä°yileÅŸtirme: %{improvement:.2f}")
    else:
        degradation = ((all_avg - selected_avg) / all_avg) * 100
        print(f"   âš ï¸ DÃ¼ÅŸÃ¼ÅŸ: %{degradation:.2f}")

# =============================================================================
# 17. SONUÃ‡LARI CSV DOSYASINA KAYDETME
# =============================================================================

print(f"\nğŸ’¾ SonuÃ§lar CSV dosyasÄ±na kaydediliyor...")

# Ana sonuÃ§lar tablosu (TÃœM METRÄ°KLER dahil)
results_df.to_csv('covertype_feature_selection_results_complete.csv', index=False)
print("âœ… 'covertype_feature_selection_results_complete.csv' dosyasÄ± oluÅŸturuldu")

# SeÃ§ilen Ã¶znitelikler tablosu
selected_features_df = pd.DataFrame({
    'Feature_Index': selected_features,
    'Feature_Name': selected_feature_names,
    'Feature_Type': ['Continuous' if i < 10 else 'Soil_Type' if i < 50 else 'Wilderness_Area'
                    for i in selected_features]
})
selected_features_df.to_csv('selected_features_fox_optimizer.csv', index=False)
print("âœ… 'selected_features_fox_optimizer.csv' dosyasÄ± oluÅŸturuldu")

# Ã–zet istatistikler
summary_stats = {
    'Total_Features': [X.shape[1]],
    'Selected_Features': [len(selected_features)],
    'Reduction_Percentage': [(1 - len(selected_features)/X.shape[1])*100],
    'Best_Fitness': [fox_optimizer.best_fitness],
    'Best_Classifier': [best_classifier_name],
    'Best_Feature_Type': [best_feature_type],
    'Best_Accuracy': [best_model_info['accuracy']],
    'Best_F1_Score': [best_model_info['f1_score']],
    'Best_Specificity': [best_model_info['specificity']],
    'Algorithm_Runtime_Seconds': [end_time - start_time]
}

summary_df = pd.DataFrame(summary_stats)
summary_df.to_csv('project_summary_complete.csv', index=False)
print("âœ… 'project_summary_complete.csv' dosyasÄ± oluÅŸturuldu")

# Metrik karÅŸÄ±laÅŸtÄ±rma tablosu
metric_comparison = []
for metric in metrics_analysis:
    all_avg = results_df[results_df['feature_type'] == 'TÃ¼m Ã–znitelikler'][metric].mean()
    selected_avg = results_df[results_df['feature_type'] == 'SeÃ§ilen Ã–znitelikler'][metric].mean()

    metric_comparison.append({
        'Metric': metric.replace('_', ' ').title(),
        'All_Features_Average': all_avg,
        'Selected_Features_Average': selected_avg,
        'Improvement_Percentage': ((selected_avg - all_avg) / all_avg) * 100 if all_avg > 0 else 0
    })

metric_comparison_df = pd.DataFrame(metric_comparison)
metric_comparison_df.to_csv('metric_comparison_analysis.csv', index=False)
print("âœ… 'metric_comparison_analysis.csv' dosyasÄ± oluÅŸturuldu")

# =============================================================================
# 18. PROJE SONUCU VE Ã–NERÄ°LER
# =============================================================================

print(f"\n" + "="*60)
print("ğŸ‰ PROJE BAÅARIYLA TAMAMLANDI!")
print("="*60)

print(f"\nğŸ“ˆ BAÅARIMLAR:")
print(f"   âœ… FOX Optimizer algoritmasÄ± baÅŸarÄ±yla uygulandÄ±")
print(f"   âœ… {X.shape[1]} Ã¶znitelikten {len(selected_features)} tanesi seÃ§ildi (%{(1-len(selected_features)/X.shape[1])*100:.1f} azalma)")
print(f"   âœ… 5 farklÄ± sÄ±nÄ±flandÄ±rÄ±cÄ± ile kapsamlÄ± test yapÄ±ldÄ±")
print(f"   âœ… TÃœM gerekli metrikler hesaplandÄ± (doÄŸruluk, hassasiyet, duyarlÄ±lÄ±k, F1-score, Ã¶zgÃ¼llÃ¼k)")
print(f"   âœ… SonuÃ§lar gÃ¶rselleÅŸtirildi ve CSV dosyalarÄ±na kaydedildi")

print(f"\nğŸ” TEKNIK BULGULAR:")
# En etkili Ã¶znitelik tÃ¼rÃ¼nÃ¼ bul
feature_type_counts = {'Continuous': len(continuous_features),
                       'Soil_Type': len(soil_features),
                       'Wilderness_Area': len(wilderness_features)}
most_effective_feature_type = max(feature_type_counts, key=feature_type_counts.get)
print(f"   â€¢ En etkili Ã¶znitelik tÃ¼rÃ¼: {most_effective_feature_type}")
print(f"   â€¢ En baÅŸarÄ±lÄ± sÄ±nÄ±flandÄ±rÄ±cÄ±: {best_classifier_name}")
print(f"   â€¢ Ã–nerilen konfigÃ¼rasyon: {best_feature_type}")

print(f"\nğŸ“Š PERFORMANS Ã–ZETÄ°:")
print(f"   â€¢ En yÃ¼ksek doÄŸruluk: {results_df['accuracy'].max():.4f}")
print(f"   â€¢ En yÃ¼ksek F1-score: {results_df['f1_score'].max():.4f}")
print(f"   â€¢ En yÃ¼ksek Ã¶zgÃ¼llÃ¼k: {results_df['specificity'].max():.4f}")
print(f"   â€¢ Ortalama eÄŸitim sÃ¼resi azalmasÄ±: %{((all_features_time - selected_features_time) / all_features_time) * 100:.1f}")

print(f"\nğŸ’¡ Ã–NERÄ°LER:")
print(f"   â€¢ FOX Optimizer algoritmasÄ± Covertype veri setinde etkili sonuÃ§lar verdi")
print(f"   â€¢ SeÃ§ilen Ã¶znitelikler ile hesaplama maliyeti azaltÄ±lÄ±rken performans korundu")
print(f"   â€¢ Gelecek Ã§alÄ±ÅŸmalarda farklÄ± doÄŸadan ilhamlÄ± algoritmalar karÅŸÄ±laÅŸtÄ±rÄ±labilir")
print(f"   â€¢ Cross-validation fold sayÄ±sÄ± artÄ±rÄ±larak daha gÃ¼venilir sonuÃ§lar elde edilebilir")

print(f"\nğŸ“‹ OLUÅTURULAN DOSYALAR:")
print(f"   ğŸ“„ covertype_feature_selection_results_complete.csv")
print(f"   ğŸ“„ selected_features_fox_optimizer.csv")
print(f"   ğŸ“„ project_summary_complete.csv")
print(f"   ğŸ“„ metric_comparison_analysis.csv")
```
