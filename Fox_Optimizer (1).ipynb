```python
# =============================================================================
# COVERTYPE VERİ SETİ ÜZERİNDE ÖZNİTELİK SEÇİMİ PROJESİ
# FOX OPTIMIZER ALGORİTMASI KULLANARAK
# =============================================================================

"""
Bu proje Covertype veri seti üzerinde doğadan ilhamlı FOX Optimizer algoritması
kullanarak öznitelik seçimi yapmayı amaçlar.

Proje Adımları:
1. Gerekli kütüphaneleri yükle
2. Veri setini indir ve incele
3. FOX Optimizer algoritmasını kodla
4. Öznitelik seçimi yap
5. Farklı sınıflandırıcılarla test et
6. Sonuçları karşılaştır ve raporla
"""

# =============================================================================
# 1. KÜTÜPHANALARI YÜKLEME
# =============================================================================

print("📦 Gerekli kütüphaneler yükleniyor...")

# Temel kütüphaneler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import time
import warnings
warnings.filterwarnings('ignore')

# Scikit-learn kütüphaneleri
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import (accuracy_score, precision_score, recall_score,
                           f1_score, confusion_matrix, classification_report)

# UCI veri seti kütüphanesi
!pip install ucimlrepo

from ucimlrepo import fetch_ucirepo

# Grafiklerin daha güzel görünmesi için
plt.style.use('default')
sns.set_palette("husl")

print("✅ Tüm kütüphaneler başarıyla yüklendi!")

# =============================================================================
# 2. VERİ SETİNİ İNDİRME VE İNCELEME
# =============================================================================

print("\n🌲 Covertype veri seti indiriliyor...")

# Covertype veri setini UCI Repository'den indir
covertype = fetch_ucirepo(id=31)

# Veri setini Pandas DataFrame'e dönüştür
X_df = covertype.data.features  # Öznitelikler
y_df = covertype.data.targets   # Hedef değişken (orman türü)

print(f"📊 Veri seti boyutu: {X_df.shape[0]} satır x {X_df.shape[1]} öznitelik")

# Veri seti hakkında bilgi
print("\n📋 Veri seti meta verileri:")
print(covertype.metadata)

print("\n📝 Öznitelik bilgileri:")
print(covertype.variables.head(10))

# İlk 5 satırı göster
print("\n👀 Verinin ilk 5 satırı:")
print(X_df.head())

print("\n🎯 Hedef değişken dağılımı:")
print(y_df.value_counts().sort_index())

# =============================================================================
# 3. VERİ ÖN İŞLEME
# =============================================================================

print("\n🔧 Veri ön işleme başlıyor...")

# Veriyi numpy array'e dönüştür
X = X_df.values
y = y_df.values.ravel()  # Tek boyutlu hale getir

print(f"Öznitelik matrisi boyutu: {X.shape}")
print(f"Hedef değişken boyutu: {y.shape}")

# Veri seti çok büyük olduğu için örnekleme yapalım (işlemi hızlandırmak için)
if X.shape[0] > 20000:
    print("⚡ Veri seti büyük, örnekleme yapılıyor...")
    np.random.seed(42)
    sample_size = 20000
    indices = np.random.choice(X.shape[0], sample_size, replace=False)
    X = X[indices]
    y = y[indices]
    print(f"Örnekleme sonrası boyut: {X.shape}")

# Sayısal öznitelikleri standartlaştır (ilk 10 öznitelik sayısal)
scaler = StandardScaler()
X_scaled = X.copy()
X_scaled[:, :10] = scaler.fit_transform(X[:, :10])

print("✅ Veri ön işleme tamamlandı!")

# Eğitim ve test kümelerini ayır
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42, stratify=y
)

print(f"Eğitim seti boyutu: {X_train.shape}")
print(f"Test seti boyutu: {X_test.shape}")

# =============================================================================
# 4. FOX OPTIMIZER ALGORİTMASINI KODLAMA
# =============================================================================

print("\n🦊 FOX Optimizer algoritması kodlanıyor...")

class FOXOptimizer:
    """
    FOX Optimizer Algoritması

    Bu algoritma tilkilerin avlanma davranışlarından esinlenir:
    - Tilkiler avlarını takip eder (exploration - keşif)
    - Avlarını kuşatır (exploitation - sömürü)
    - En iyi avı bulana kadar bu stratejileri tekrarlar

    Öznitelik seçimi için:
    - Her tilki bir öznitelik alt kümesini temsil eder
    - En iyi sınıflandırma performansını veren alt kümeyi arar
    """

    def __init__(self, n_foxes=15, max_iter=15, p_pursuit=0.3, p_encircle=0.4):
        """
        Parametreler:
        - n_foxes: Tilki sayısı (popülasyon büyüklüğü)
        - max_iter: Maksimum iterasyon sayısı
        - p_pursuit: Takip etme olasılığı
        - p_encircle: Kuşatma olasılığı
        """
        self.n_foxes = n_foxes
        self.max_iter = max_iter
        self.p_pursuit = p_pursuit
        self.p_encircle = p_encircle
        self.best_solution = None
        self.best_fitness = 0.0
        self.best_subset = None
        self.fitness_history = []

    def initialize_population(self, n_features):
        """
        Tilki popülasyonunu başlat
        Her tilki binary bir vektör (0: öznitelik seçilmedi, 1: seçildi)
        """
        population = np.random.randint(0, 2, size=(self.n_foxes, n_features))

        # Her tilkinin en az bir öznitelik seçtiğinden emin ol
        for i in range(self.n_foxes):
            if np.sum(population[i]) == 0:
                population[i, np.random.randint(0, n_features)] = 1

        return population

    def calculate_fitness(self, solution, X, y):
        """
        Bir çözümün fitness değerini hesapla

        Fitness = Sınıflandırma Doğruluğu - Öznitelik Cezası

        Bu şekilde hem yüksek doğruluk hem de az öznitelik kullanmayı hedefleriz
        """
        # Hiç öznitelik seçilmediyse 0 döndür
        if np.sum(solution) == 0:
            return 0.0

        # Seçilen öznitelikleri bul
        selected_features = np.where(solution == 1)[0]

        # Seçilen özniteliklerle veri alt kümesi oluştur
        X_selected = X[:, selected_features]

        # K-NN sınıflandırıcı ile çapraz doğrulama yap
        knn = KNeighborsClassifier(n_neighbors=5)
        # Using a smaller cv value to reduce execution time
        scores = cross_val_score(knn, X_selected, y, cv=2, scoring='accuracy')

        # Ortalama doğruluk
        accuracy = np.mean(scores)

        # Öznitelik sayısı cezası (çok fazla öznitelik seçmeyi önlemek için)
        feature_ratio = np.sum(solution) / len(solution)
        penalty = 0.01 * feature_ratio

        fitness = accuracy - penalty
        return max(0, fitness)  # Negatif değer olmasını önle

    def pursue_prey(self, fox, best_fox):
        """
        Tilkinin en iyi tilkiyi takip etmesi (keşif fazı)
        """
        new_fox = fox.copy()
        for i in range(len(fox)):
            if np.random.random() < self.p_pursuit:
                new_fox[i] = best_fox[i]
        return new_fox

    def encircle_prey(self, fox):
        """
        Tilkinin mevcut pozisyonunu değiştirmesi (sömürü fazı)
        """
        new_fox = fox.copy()
        for i in range(len(fox)):
            if np.random.random() < self.p_encircle:
                new_fox[i] = 1 - new_fox[i]  # Bit çevirme

        # En az bir öznitelik seçildiğinden emin ol
        if np.sum(new_fox) == 0:
            new_fox[np.random.randint(0, len(new_fox))] = 1

        return new_fox

    def fit(self, X, y):
        """
        FOX Optimizer algoritmasını çalıştır
        """
        print(f"🦊 FOX Optimizer başlıyor: {self.n_foxes} tilki, {self.max_iter} iterasyon")

        n_features = X.shape[1]

        # 1. Popülasyonu başlat
        population = self.initialize_population(n_features)
        fitness_values = np.zeros(self.n_foxes)

        # 2. İlk fitness değerlerini hesapla
        print("🧮 İlk fitness değerleri hesaplanıyor...")
        for i in range(self.n_foxes):
            fitness_values[i] = self.calculate_fitness(population[i], X, y)

        # 3. En iyi tilkiyi bul
        best_idx = np.argmax(fitness_values)
        self.best_solution = population[best_idx].copy()
        self.best_fitness = fitness_values[best_idx]

        print(f"İlk en iyi fitness: {self.best_fitness:.4f}")

        # 4. Ana algoritma döngüsü
        for iteration in range(self.max_iter):
            print(f"\n🔄 İterasyon {iteration + 1}/{self.max_iter}")

            # Her tilki için avlanma stratejisi uygula
            for i in range(self.n_foxes):
                current_fox = population[i].copy()
                current_fitness = fitness_values[i]

                # Takip etme stratejisi
                fox_pursuit = self.pursue_prey(current_fox, self.best_solution)
                fitness_pursuit = self.calculate_fitness(fox_pursuit, X, y)

                # Kuşatma stratejisi
                fox_encircle = self.encircle_prey(current_fox)
                fitness_encircle = self.calculate_fitness(fox_encircle, X, y)

                # En iyi stratejiyi seç
                if fitness_pursuit > max(fitness_encircle, current_fitness):
                    population[i] = fox_pursuit
                    fitness_values[i] = fitness_pursuit
                elif fitness_encircle > current_fitness:
                    population[i] = fox_encircle
                    fitness_values[i] = fitness_encircle

                # Global en iyi çözümü güncelle
                if fitness_values[i] > self.best_fitness:
                    self.best_solution = population[i].copy()
                    self.best_fitness = fitness_values[i]

            # İlerleme bilgisi
            selected_count = np.sum(self.best_solution)
            avg_fitness = np.mean(fitness_values)

            print(f"En iyi fitness: {self.best_fitness:.4f}")
            print(f"Ortalama fitness: {avg_fitness:.4f}")
            print(f"Seçilen öznitelik sayısı: {selected_count}/{n_features}")

            # Fitness geçmişini kaydet
            self.fitness_history.append(self.best_fitness)

        # 5. Son sonuçları kaydet
        self.best_subset = np.where(self.best_solution == 1)[0]

        print(f"\n✅ FOX Optimizer tamamlandı!")
        print(f"Final en iyi fitness: {self.best_fitness:.4f}")
        print(f"Seçilen öznitelik sayısı: {len(self.best_subset)}")

        return self

    def transform(self, X):
        """
        Veriyi seçilen özniteliklerle dönüştür
        """
        if self.best_subset is None:
            raise ValueError("Önce fit() metodunu çağırın!")
        return X[:, self.best_subset]

    def plot_convergence(self):
        """
        Algoritmanın yakınsama grafiğini çiz
        """
        plt.figure(figsize=(10, 6))
        plt.plot(range(1, len(self.fitness_history) + 1), self.fitness_history,
                'b-o', linewidth=2, markersize=6)
        plt.title('FOX Optimizer Yakınsama Grafiği', fontsize=14, fontweight='bold')
        plt.xlabel('İterasyon', fontsize=12)
        plt.ylabel('En İyi Fitness Değeri', fontsize=12)
        plt.grid(True, alpha=0.3)
        plt.tight_layout()
        plt.show()

print("✅ FOX Optimizer sınıfı başarıyla tanımlandı!")

# =============================================================================
# 5. FOX OPTIMIZER İLE ÖZNİTELİK SEÇİMİ
# =============================================================================

print("\n🎯 FOX Optimizer ile öznitelik seçimi başlıyor...")

# FOX Optimizer nesnesini oluştur
fox_optimizer = FOXOptimizer(
    n_foxes=15,        # 15 tilki kullan
    max_iter=15,       # 15 iterasyon
    p_pursuit=0.3,     # %30 takip etme olasılığı
    p_encircle=0.4     # %40 kuşatma olasılığı
)

# Algoritmanın çalışma zamanını ölç
start_time = time.time()

# Öznitelik seçimini yap
fox_optimizer.fit(X_train, y_train)

end_time = time.time()
print(f"\n⏱️ Toplam çalışma süresi: {end_time - start_time:.2f} saniye")

# Seçilen öznitelikleri göster
selected_features = fox_optimizer.best_subset
print(f"\n🎯 Seçilen öznitelikler ({len(selected_features)} adet):")
print(f"İndeksler: {selected_features}")

# Öznitelik isimlerini göster
feature_names = X_df.columns.tolist()
selected_feature_names = [feature_names[i] for i in selected_features]
print(f"\nÖznitelik isimleri:")
for i, name in enumerate(selected_feature_names):
    print(f"{i+1:2d}. {name}")

# Yakınsama grafiğini çiz
fox_optimizer.plot_convergence()

# =============================================================================
# 6. SEÇİLEN ÖZNİTELİKLERİ GÖRSELLEŞTİRME
# =============================================================================

print("\n📊 Seçilen öznitelikler görselleştiriliyor...")

# Öznitelik seçimi sonuçlarını görselleştir
plt.figure(figsize=(15, 8))

# Tüm öznitelikler için binary gösterim
feature_selection = np.zeros(len(feature_names))
feature_selection[selected_features] = 1

# Bar grafiği
bars = plt.bar(range(len(feature_names)), feature_selection,
               color=['green' if x == 1 else 'lightgray' for x in feature_selection])

plt.title('FOX Optimizer ile Seçilen Öznitelikler', fontsize=16, fontweight='bold')
plt.xlabel('Öznitelik İndeksi', fontsize=12)
plt.ylabel('Seçildi (1) / Seçilmedi (0)', fontsize=12)
plt.ylim(-0.1, 1.1)

# Seçilen öznitelikleri vurgula
for i, bar in enumerate(bars):
    if feature_selection[i] == 1:
        bar.set_color('darkgreen')
        bar.set_alpha(0.8)

        # Öznitelik ismini yazdır
        plt.text(i, 1.05, f'{i}', ha='center', va='bottom',
                fontsize=8, rotation=45, color='darkgreen', fontweight='bold')

plt.grid(True, alpha=0.3, axis='y')
plt.tight_layout()
plt.show()

# Seçilen öznitelik türlerinin dağılımı
print("\n📈 Seçilen öznitelik türleri:")

# Öznitelik türlerini kategorize et
continuous_features = []  # Sürekli öznitelikler (ilk 10)
wilderness_features = []  # Wilderness Area (50-53)
soil_features = []        # Soil Type (10-49)

for idx in selected_features:
    if idx < 10:
        continuous_features.append(feature_names[idx])
    elif 10 <= idx < 50:
        soil_features.append(feature_names[idx])
    else:
        wilderness_features.append(feature_names[idx])

print(f"Sürekli öznitelikler: {len(continuous_features)} adet")
print(f"Toprak tipi öznitelikleri: {len(soil_features)} adet")
print(f"Wilderness area öznitelikleri: {len(wilderness_features)} adet")

# Pasta grafiği
categories = ['Sürekli', 'Toprak Tipi', 'Wilderness Area']
sizes = [len(continuous_features), len(soil_features), len(wilderness_features)]
colors = ['lightblue', 'lightcoral', 'lightgreen']

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=categories, colors=colors, autopct='%1.1f%%',
        startangle=90, textprops={'fontsize': 12})
plt.title('Seçilen Özniteliklerin Tür Dağılımı', fontsize=14, fontweight='bold')
plt.axis('equal')
plt.show()

# =============================================================================
# 7. SINIFLANDIRICILAR İLE PERFORMANS DEĞERLENDİRMESİ
# =============================================================================

print("\n🤖 Farklı sınıflandırıcılarla performans değerlendirmesi...")

# Seçilen özniteliklerle eğitim ve test verilerini dönüştür
X_train_selected = fox_optimizer.transform(X_train)
X_test_selected = fox_optimizer.transform(X_test)

print(f"Orijinal öznitelik sayısı: {X_train.shape[1]}")
print(f"Seçilen öznitelik sayısı: {X_train_selected.shape[1]}")

# Sınıflandırıcıları tanımla
classifiers = {
    'K-Nearest Neighbors': KNeighborsClassifier(n_neighbors=5),
    'Decision Tree': DecisionTreeClassifier(random_state=42, max_depth=10),
    'Naive Bayes': GaussianNB(),
    'Random Forest': RandomForestClassifier(n_estimators=50, random_state=42, max_depth=10),
    'Support Vector Machine': SVC(kernel='rbf', random_state=42, C=1.0)
}

def evaluate_classifier(clf, X_train, X_test, y_train, y_test, name, feature_type):
    """
    Sınıflandırıcının performansını değerlendir
    """
    print(f"\n--- {name} ({feature_type}) ---")

    # Eğitim süresi ölçümü
    start_time = time.time()
    clf.fit(X_train, y_train)
    training_time = time.time() - start_time

    # Tahmin
    y_pred = clf.predict(X_test)

    # Performans metrikleri
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_test, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_test, y_pred, average='weighted', zero_division=0)

    print(f"⏱️  Eğitim süresi: {training_time:.3f} saniye")
    print(f"🎯 Doğruluk (Accuracy): {accuracy:.4f}")
    print(f"🔍 Hassasiyet (Precision): {precision:.4f}")
    print(f"📊 Duyarlılık (Recall): {recall:.4f}")
    print(f"⚖️  F1-Score: {f1:.4f}")

    return {
        'classifier': name,
        'feature_type': feature_type,
        'accuracy': accuracy,
        'precision': precision,
        'recall': recall,
        'f1_score': f1,
        'training_time': training_time,
        'n_features': X_train.shape[1]
    }

# Sonuçları saklamak için liste
results = []

print("\n" + "="*60)
print("PERFORMANS DEĞERLENDİRMESİ")
print("="*60)

# Her sınıflandırıcı için hem tüm öznitelikler hem de seçilen öznitelikler ile test et
for name, clf in classifiers.items():

    # Tüm özniteliklerle test et
    result_all = evaluate_classifier(
        clf, X_train, X_test, y_train, y_test, name, "Tüm Öznitelikler"
    )
    results.append(result_all)

    # Seçilen özniteliklerle test et
    result_selected = evaluate_classifier(
        clf, X_train_selected, X_test_selected, y_train, y_test, name, "Seçilen Öznitelikler"
    )
    results.append(result_selected)

# =============================================================================
# 8. SONUÇLARI KARŞILAŞTIRMA VE GÖRSELLEŞTİRME
# =============================================================================

print("\n📊 Sonuçlar karşılaştırılıyor ve görselleştiriliyor...")

# Sonuçları DataFrame'e dönüştür
results_df = pd.DataFrame(results)

# Sonuçları pivot table olarak düzenle
metrics = ['accuracy', 'precision', 'recall', 'f1_score', 'training_time']

# Her metrik için karşılaştırma grafiği
fig, axes = plt.subplots(2, 3, figsize=(18, 12))
axes = axes.ravel()

classifier_names = list(classifiers.keys())
n_classifiers = len(classifier_names)

for i, metric in enumerate(metrics):
    ax = axes[i]

    # Tüm öznitelikler için değerler
    all_features_values = []
    selected_features_values = []

    for clf_name in classifier_names:
        all_val = results_df[(results_df['classifier'] == clf_name) &
                           (results_df['feature_type'] == 'Tüm Öznitelikler')][metric].values[0]
        selected_val = results_df[(results_df['classifier'] == clf_name) &
                                (results_df['feature_type'] == 'Seçilen Öznitelikler')][metric].values[0]

        all_features_values.append(all_val)
        selected_features_values.append(selected_val)

    # Bar grafiği
    x = np.arange(n_classifiers)
    width = 0.35

    bars1 = ax.bar(x - width/2, all_features_values, width,
                   label='Tüm Öznitelikler', alpha=0.8, color='lightblue')
    bars2 = ax.bar(x + width/2, selected_features_values, width,
                   label='Seçilen Öznitelikler', alpha=0.8, color='lightcoral')

    # Grafik düzenlemeleri
    ax.set_xlabel('Sınıflandırıcı')
    ax.set_ylabel(metric.replace('_', ' ').title())
    ax.set_title(f'{metric.replace("_", " ").title()} Karşılaştırması')
    ax.set_xticks(x)
    ax.set_xticklabels([name.split()[0] for name in classifier_names], rotation=45)
    ax.legend()
    ax.grid(True, alpha=0.3)

    # Değerleri bar üzerinde göster
    for bar in bars1:
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.3f}', ha='center', va='bottom', fontsize=8)

    for bar in bars2:
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{height:.3f}', ha='center', va='bottom', fontsize=8)

# Son subplot'u kaldır
axes[-1].remove()

plt.tight_layout()
plt.show()

# =============================================================================
# 9. ÖZET RAPOR
# =============================================================================

print("\n" + "="*60)
print("PROJE ÖZET RAPORU")
print("="*60)

print(f"\n📊 VERİ SETİ BİLGİLERİ:")
print(f"   • Toplam örnek sayısı: {len(y)}")
print(f"   • Toplam öznitelik sayısı: {X.shape[1]}")
print(f"   • Sınıf sayısı: {len(np.unique(y))}")

print(f"\n🦊 FOX OPTIMIZER SONUÇLARI:")
print(f"   • Seçilen öznitelik sayısı: {len(selected_features)}")
print(f"   • Öznitelik azaltma oranı: %{(1 - len(selected_features)/X.shape[1])*100:.1f}")
print(f"   • En iyi fitness değeri: {fox_optimizer.best_fitness:.4f}")

print(f"\n🏆 EN İYİ PERFORMANS GÖSTERENLERİ:")

# En iyi doğruluk
best_accuracy = results_df.loc[results_df['accuracy'].idxmax()]
print(f"\n   📍 En Yüksek Doğruluk:")
print(f"      {best_accuracy['classifier']} ({best_accuracy['feature_type']})")
print(f"      Doğruluk: {best_accuracy['accuracy']:.4f}")

# En iyi F1-score
best_f1 = results_df.loc[results_df['f1_score'].idxmax()]
print(f"\n   📍 En Yüksek F1-Score:")
print(f"      {best_f1['classifier']} ({best_f1['feature_type']})")
print(f"      F1-Score: {best_f1['f1_score']:.4f}")

# En hızlı eğitim
fastest_training = results_df.loc[results_df['training_time'].idxmin()]
print(f"\n   📍 En Hızlı Eğitim:")
print(f"      {fastest_training['classifier']} ({fastest_training['feature_type']})")
print(f"      Süre: {fastest_training['training_time']:.3f} saniye")

print(f"\n💡 ÖNEMLİ BULGULAR:")

# Seçilen özniteliklerle performans karşılaştırması
all_features_avg = results_df[results_df['feature_type'] == 'Tüm Öznitelikler']['accuracy'].mean()
selected_features_avg = results_df[results_df['feature_type'] == 'Seçilen Öznitelikler']['accuracy'].mean()

print(f"   • Tüm özniteliklerle ortalama doğruluk: {all_features_avg:.4f}")
print(f"   • Seçilen özniteliklerle ortalama doğruluk: {selected_features_avg:.4f}")

if selected_features_avg > all_features_avg:
    print(f"   ✅ FOX Optimizer ile seçilen öznitelikler daha iyi performans gösterdi!")
    print(f"      İyileştirme: %{((selected_features_avg - all_features_avg) / all_features_avg) * 100:.2f}")
else:
    print(f"   ⚠️  Öznitelik seçimi performansı biraz düşürdü, ancak hesaplama maliyetini azalttı.")

# Eğitim süresi karşılaştırması
all_features_time = results_df[results_df['feature_type'] == 'Tüm Öznitelikler']['training_time'].mean()
selected_features_time = results_df[results_df['feature_type'] == 'Seçilen Öznitelikler']['training_time'].mean()

print(f"\n   ⏱️  Ortalama eğitim süreleri:")
print(f"      Tüm öznitelikler: {all_features_time:.3f} saniye")
print(f"      Seçilen öznitelikler: {selected_features_time:.3f} saniye")
print(f"      Hızlanma: %{((all_features_time - selected_features_time) / all_features_time) * 100:.1f}")

# =============================================================================
# 10. DETAYLI SONUÇ TABLOSU
# =============================================================================

print(f"\n📋 DETAYLI SONUÇ TABLOSU:")
print("="*60)

# Sonuçları düzenli tablo halinde göster
display_df = results_df.round(4)
display_df = display_df.sort_values(['classifier', 'feature_type'])

# Sütun isimlerini Türkçeleştir
display_df_turkish = display_df.copy()
display_df_turkish.columns = ['Sınıflandırıcı', 'Öznitelik Türü', 'Doğruluk',
                             'Hassasiyet', 'Duyarlılık', 'F1-Puanı',
                             'Eğitim Süresi', 'Öznitelik Sayısı']

print(display_df_turkish.to_string(index=False))

# =============================================================================
# 11. SEÇİLEN ÖZNİTELİKLERİN DETAYLI ANALİZİ
# =============================================================================

print(f"\n🔍 SEÇİLEN ÖZNİTELİKLERİN DETAYLI ANALİZİ:")
print("="*60)

# Seçilen özniteliklerin istatistikleri
print(f"\n📊 Seçilen Öznitelik İstatistikleri:")
selected_data = X_train_selected
feature_stats = []

for i, feature_idx in enumerate(selected_features):
    feature_name = feature_names[feature_idx]
    feature_data = X_train[:, feature_idx]

    stats = {
        'İndeks': feature_idx,
        'İsim': feature_name,
        'Ortalama': np.mean(feature_data),
        'Std Sapma': np.std(feature_data),
        'Min': np.min(feature_data),
        'Max': np.max(feature_data)
    }
    feature_stats.append(stats)

feature_stats_df = pd.DataFrame(feature_stats)
print(feature_stats_df.round(3).to_string(index=False))

# =============================================================================
# 12. GÖRSELLEŞTİRME: PERFORMANS RADAR CHART
# =============================================================================

print(f"\n📊 Radar chart ile performans görselleştirmesi...")

# En iyi performans gösteren sınıflandırıcıları seç
top_classifiers = ['Random Forest', 'K-Nearest Neighbors', 'Decision Tree']

# Radar chart için veri hazırla
metrics_for_radar = ['accuracy', 'precision', 'recall', 'f1_score']
angles = np.linspace(0, 2 * np.pi, len(metrics_for_radar), endpoint=False).tolist()
angles += angles[:1]  # Döngüyü tamamla

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7), subplot_kw=dict(projection='polar'))

colors = ['blue', 'red', 'green']

# Tüm öznitelikler için radar chart
ax1.set_title('Tüm Öznitelikler ile Performans', y=1.08, fontsize=14, fontweight='bold')
for i, clf_name in enumerate(top_classifiers):
    values = []
    for metric in metrics_for_radar:
        value = results_df[(results_df['classifier'] == clf_name) &
                          (results_df['feature_type'] == 'Tüm Öznitelikler')][metric].values[0]
        values.append(value)
    values += values[:1]  # Döngüyü tamamla

    ax1.plot(angles, values, 'o-', linewidth=2, label=clf_name, color=colors[i])
    ax1.fill(angles, values, alpha=0.25, color=colors[i])

ax1.set_xticks(angles[:-1])
ax1.set_xticklabels([m.replace('_', '\n').title() for m in metrics_for_radar])
ax1.set_ylim(0, 1)
ax1.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
ax1.grid(True)

# Seçilen öznitelikler için radar chart
ax2.set_title('Seçilen Öznitelikler ile Performans', y=1.08, fontsize=14, fontweight='bold')
for i, clf_name in enumerate(top_classifiers):
    values = []
    for metric in metrics_for_radar:
        value = results_df[(results_df['classifier'] == clf_name) &
                          (results_df['feature_type'] == 'Seçilen Öznitelikler')][metric].values[0]
        values.append(value)
    values += values[:1]  # Döngüyü tamamla

    ax2.plot(angles, values, 'o-', linewidth=2, label=clf_name, color=colors[i])
    ax2.fill(angles, values, alpha=0.25, color=colors[i])

ax2.set_xticks(angles[:-1])
ax2.set_xticklabels([m.replace('_', '\n').title() for m in metrics_for_radar])
ax2.set_ylim(0, 1)
ax2.legend(loc='upper right', bbox_to_anchor=(1.3, 1.0))
ax2.grid(True)

plt.tight_layout()
plt.show()

# =============================================================================
# 13. CONFUSION MATRIX GÖRSELLEŞTİRMESİ
# =============================================================================

print(f"\n🎯 En iyi performans gösteren modelin confusion matrix'i...")

# En iyi F1-score'a sahip modeli bul
best_model_info = results_df.loc[results_df['f1_score'].idxmax()]
best_classifier_name = best_model_info['classifier']
best_feature_type = best_model_info['feature_type']

print(f"En iyi model: {best_classifier_name} ({best_feature_type})")

# En iyi modeli yeniden eğit ve confusion matrix oluştur
if best_feature_type == 'Seçilen Öznitelikler':
    X_train_best = X_train_selected
    X_test_best = X_test_selected
else:
    X_train_best = X_train
    X_test_best = X_test

# Modeli seç ve eğit
if 'Random Forest' in best_classifier_name:
    best_clf = RandomForestClassifier(n_estimators=50, random_state=42, max_depth=10)
elif 'K-Nearest' in best_classifier_name:
    best_clf = KNeighborsClassifier(n_neighbors=5)
elif 'Decision Tree' in best_classifier_name:
    best_clf = DecisionTreeClassifier(random_state=42, max_depth=10)
elif 'Naive Bayes' in best_classifier_name:
    best_clf = GaussianNB()
else:  # SVM
    best_clf = SVC(kernel='rbf', random_state=42, C=1.0)

best_clf.fit(X_train_best, y_train)
y_pred_best = best_clf.predict(X_test_best)

# Confusion matrix
cm = confusion_matrix(y_test, y_pred_best)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=[f'Sınıf {i}' for i in range(1, 8)],
            yticklabels=[f'Sınıf {i}' for i in range(1, 8)])
plt.title(f'Confusion Matrix - {best_classifier_name}\n({best_feature_type})',
          fontsize=14, fontweight='bold')
plt.xlabel('Tahmin Edilen Sınıf', fontsize=12)
plt.ylabel('Gerçek Sınıf', fontsize=12)
plt.tight_layout()
plt.show()

# Sınıf bazında performans raporu
print(f"\n📊 Sınıf bazında detaylı performans raporu:")
class_report = classification_report(y_test, y_pred_best,
                                   target_names=[f'Sınıf {i}' for i in range(1, 8)])
print(class_report)

# =============================================================================
# 14. SONUÇLARI CSV DOSYASINA KAYDETME
# =============================================================================

print(f"\n💾 Sonuçlar CSV dosyasına kaydediliyor...")

# Ana sonuçlar tablosu
results_df.to_csv('covertype_feature_selection_results.csv', index=False)
print("✅ 'covertype_feature_selection_results.csv' dosyası oluşturuldu")

# Seçilen öznitelikler tablosu
selected_features_df = pd.DataFrame({
    'Feature_Index': selected_features,
    'Feature_Name': selected_feature_names,
    'Feature_Type': ['Continuous' if i < 10 else 'Soil_Type' if i < 50 else 'Wilderness_Area'
                    for i in selected_features]
})
selected_features_df.to_csv('selected_features_fox_optimizer.csv', index=False)
print("✅ 'selected_features_fox_optimizer.csv' dosyası oluşturuldu")

# Özet istatistikler
summary_stats = {
    'Total_Features': [X.shape[1]],
    'Selected_Features': [len(selected_features)],
    'Reduction_Percentage': [(1 - len(selected_features)/X.shape[1])*100],
    'Best_Fitness': [fox_optimizer.best_fitness],
    'Best_Classifier': [best_classifier_name],
    'Best_Feature_Type': [best_feature_type],
    'Best_Accuracy': [best_model_info['accuracy']],
    'Best_F1_Score': [best_model_info['f1_score']],
    'Algorithm_Runtime_Seconds': [end_time - start_time]
}

summary_df = pd.DataFrame(summary_stats)
summary_df.to_csv('project_summary.csv', index=False)
print("✅ 'project_summary.csv' dosyası oluşturuldu")

# =============================================================================
# 15. PROJE SONUCU VE ÖNERİLER
# =============================================================================

print(f"\n" + "="*60)
print("🎉 PROJE BAŞARIYLA TAMAMLANDI!")
print("="*60)

print(f"\n📈 BAŞARIMLAR:")
print(f"   ✅ FOX Optimizer algoritması başarıyla uygulandı")
print(f"   ✅ {X.shape[1]} öznitelikten {len(selected_features)} tanesi seçildi (%{(1-len(selected_features)/X.shape[1])*100:.1f} azalma)")
print(f"   ✅ 5 farklı sınıflandırıcı ile kapsamlı test yapıldı")
print(f"   ✅ Sonuçlar görselleştirildi ve CSV dosyalarına kaydedildi")

print(f"\n🔍 TEKNIK BULGULAR:")
# Corrected line to determine the most effective feature type
feature_type_counts = {'Continuous': len(continuous_features),
                       'Soil_Type': len(soil_features),
                       'Wilderness_Area': len(wilderness_features)}
most_effective_feature_type = max(feature_type_counts, key=feature_type_counts.get)
print(f"   • En etkili öznitelik türü: {most_effective_feature_type}")
print(f"   • En başarılı sınıflandırıcı: {best_classifier_name}")
print(f"   • Önerilen konfigürasyon: {best_feature_type}")
```
